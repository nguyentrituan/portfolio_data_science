```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo một dãy số mẫu về doanh số bán hàng hàng ngày
np.random.seed(0)
sales_data = np.random.randint(50, 150, 365)


# Hàm tính auto-correlation
def auto_correlation(data, lag=1):
    n = len(data)
    mean = np.mean(data)
    numerator = np.sum((data[:n - lag] - mean) * (data[lag:] - mean))
    denominator = np.sum((data - mean) ** 2)

    auto_corr = numerator / denominator if denominator != 0 else 0
    return auto_corr

# # Tính toán auto-correlation cho mỗi độ trễ từ 1 đến 30
lags = range(1, 31)
correlation_values = [auto_correlation(sales_data, lag) for lag in lags]

# Vẽ biểu đồ để hiển thị auto-correlation
plt.figure(figsize=(10, 5))
plt.bar(lags, correlation_values)
plt.title('Auto-correlation of Daily Sales Data')
plt.xlabel('Lag (Days)')
plt.ylabel('Auto-correlation Coefficient')
plt.show()
```
